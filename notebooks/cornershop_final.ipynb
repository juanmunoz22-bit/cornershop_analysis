```python
import haversine as hs
```

```python
def distance_from(loc1,loc2): 
    dist=hs.haversine(loc1,loc2)
    return round(dist,2)
```

```python
import folium
```

```python
def create_map(df: pd.DataFrame):
  m = folium.Map(
      location=[df.store_loc.values[0][0], df.store_loc.values[0][1]], 
      zoom_start=12, 
      tiles='OpenStreetMap',
      width='75%', 
      height='75%',
    )

  for _, row in df.iterrows():
    if row['cond'] == 'out':
      cluster_colour = 'red'
    else:
      cluster_colour = 'green'
    
    folium.CircleMarker(
        location=[row.order_loc[0], row.order_loc[1]],
        radius=5,
        color=cluster_colour,
        fill=True,
        fill_color=cluster_colour
    ).add_to(m)
    
    folium.Marker(
        location=[row.store_loc[0], row.store_loc[1]],
    ).add_to(m)

    folium.Circle(
        location=[row.store_loc[0], row.store_loc[1]],
        radius=3000,
        color='black',
        fill=False,
    ).add_to(m)

  return m
```

```python
def is_in(dist):
  cond = 'out'
  if dist < 3:
    cond = 'in'
    return cond
  return cond
```

```python
df_specific_store['cond'] = 'out'
df_specific_store.loc[df_specific_store['dist']<3,'cond'] = 'in'
df_specific_store
```
