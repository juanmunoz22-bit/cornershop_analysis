```python
import haversine as hs
```

```python
def distance_from(loc1,loc2): 
    dist=hs.haversine(loc1,loc2)
    return round(dist,2)
```

```python
df_order_store['order_loc'] = list(zip(df_order_store['order_lat'], df_order_store['order_lng']))
df_order_store['store_loc'] = list(zip(df_order_store['store_lat'], df_order_store['store_lng']))
df_order_store
```

```python
df_order_store['order_loc'][0]
```

```python
df_order_store['dist'] = df_order_store.apply(lambda x: distance_from(x['order_loc'], x['store_loc']), axis=1)
df_order_store
```
